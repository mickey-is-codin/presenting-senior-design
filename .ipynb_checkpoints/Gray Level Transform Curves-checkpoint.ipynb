```python
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
from pylab import *

%matplotlib inline
```

```python
board_im = Image.open('BookFigureImages/Fig0114acktboard.tif')
board_array = np.array(board_im.convert('L'))
board_gray_im = Image.fromarray(board_array)

print board_array.shape
print board_array.dtype
```

```python
fig = plt.figure(figsize=(15,10))

subplot(1,2,1)
imshow(board_im)

subplot(1,2,2)
imshow(board_gray_im)
```

```python
# Invert Image

inverted_array = 255 - board_array
inverted_im = Image.fromarray(inverted_array)

fig = plt.figure(figsize=(15,10))

subplot(1,2,1)
imshow(board_gray_im)

subplot(1,2,2)
imshow(inverted_im)
```

```python
# Clamp Image

clamped_array = (100.0/255) * board_array + 100
clamped_im = Image.fromarray(clamped_array)

fig = plt.figure(figsize=(15,10))

subplot(1,2,1)
imshow(board_gray_im)

subplot(1,2,2)
imshow(clamped_im)
```

```python
# Square Image

squared_array = 255.0 * (board_array/255.0)**2
squared_im = Image.fromarray(squared_array)

fig = plt.figure(figsize=(15,10))

subplot(1,2,1)
imshow(board_gray_im)

subplot(1,2,2)
imshow(squared_im)
```

```python
# Transform curves

fig= plt.figure(figsize=(5,5))
ax = fig.add_subplot(111)

x = np.arange(256)

y = x
y1 = 255 - x
y2 = (100.0/255) * x + 100
y3 = 255.0 * (x/255.0)**2

plt.plot(y)
plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

ax.legend(('y = x','y1 = 255 - x','y2 = (100.0/255) * x + 100','y3 = 255.0 * (x/255.0)**2'))

plt.show()
```
